```python
bins = [0, 17, 64, 200]  # age ranges
labels = ["child", "adult", "senior"]
df["age_group"] = pd.cut(df["Age"], bins=bins, labels=labels, right=True)
```

```python
total_passengers = len(df)
age_group_counts = df.groupby("age_group")["PassengerId"].count()
print("Passenger counts by age group: \n",age_group_counts)
```

```python
age_group_percentage = (age_group_counts / total_passengers) * 100
print("\nProportion of passengers by age group:",age_group_percentage)
```
