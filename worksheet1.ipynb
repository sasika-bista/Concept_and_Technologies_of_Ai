```python
#problem4
A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

#1
inverseA = np.linalg.inv(A)
print(A.dot(inverseA))

```

```python
#2
AB =A.dot(B)
BA= B.dot(A)
if( np.array_equal(AB,BA) ):
    print("AB = BA")
else:
    print("AB ≠ BA")

```

```python
#3
LHS = (A.dot(B)).T
RHS = B.T.dot(A.T)
if( np.array_equal(LHS, RHS) ):
    print("(AB)ᵀ = Bᵀ Aᵀ")
else:
    print("(AB)ᵀ ≠ Bᵀ Aᵀ")
```

```python
#experiment
#1
# Python lists
list1 = list(range(1000000))
list2 = list(range(1000000))
pyStart = time.time()
list_add = [list1[i] + list2[i] for i in range(1000000)]
pyEnd = time.time()
print("Python List Addition Time:", pyEnd - pyStart, "seconds")

# NumPy arrays
arr1 = np.arange(1000000)
arr2 = np.arange(1000000)
npStart = time.time()
np_add = arr1 + arr2
npEnd = time.time()
print("NumPy Addition Time:", npEnd - npStart, "seconds")
```
