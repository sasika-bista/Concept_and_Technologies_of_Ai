```python
#problem 2
df = pd.read_csv("/content/drive/MyDrive/Dataset/medical_students_dataset.csv")
df.head()
```

```python
df.info()
null_col = df.isnull().sum()
print("null count:",null_col)
```

```python

#3
missing_cols = null_col[null_col > 0].index
print("\nColumns with missing values:", list(missing_cols))

for col in missing_cols:
    if df[col].dtype == "object":
        df[col] = df[col].fillna(df[col].mode()[0])   # categorical
        print(f"{col}: filled with MODE →", df[col].mode()[0])
    else:
        df[col] = df[col].fillna(df[col].median())    # numeric
        print(f"{col}: filled with MEDIAN →", df[col].median())
```

```python
#4
duplicates = df.duplicated().sum()

df = df.drop_duplicates()
df.info()
```

```python
#3.2
#poblem 1
df = pd.read_csv("/content/drive/MyDrive/Dataset/Titanic-Dataset.csv")
df.info()
```

```python
#creating subset df
subset_df = df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
ones_df = subset_df[subset_df['Pclass'] == 1]
```

```python

mean_fare = ones_df['Fare'].mean()
median_fare = ones_df['Fare'].median()
max_fare = ones_df['Fare'].max()
min_fare = ones_df['Fare'].min()

print("Mean Fare:", mean_fare)
print("Median Fare:", median_fare)
print("Maximum Fare:", max_fare)
print("Minimum Fare:", min_fare)
```

```python
ones_df['Fare'].describe()
```

```python
#problem 3
#1
# Create binary columns for each category
embarked_dummies = pd.get_dummies(df['Embarked'], prefix='Embarked')
```

```python
#2
df = pd.concat([df, embarked_dummies], axis=1)
```

```python
#3
df = df.drop(columns=['Embarked'])
```

```python
#4
print(df.head())
```

```python
#problem 4
import matplotlib.pyplot as plt

mean_survival_by_sex = df.groupby('Sex')['Survived'].mean()
print(mean_survival_by_sex)
```

```python
# Create the plot
plt.bar(mean_survival_by_sex.index, mean_survival_by_sex.values, color=['lightblue', 'lightpink'])

# Add title and labels
plt.title("Mean Survival Rate by Gender")
plt.xlabel("Gender")
plt.ylabel("Mean Survival Rate")

# Show the values on top of bars
for i, v in enumerate(mean_survival_by_sex.values):
    plt.text(i, v + 0.02, f"{v:.2f}", ha='center', fontweight='bold')

plt.ylim(0, 1)  # Set y-axis from 0 to 1
plt.show()
```

```python
#problem 5
df = pd.read_csv("/content/drive/MyDrive/Dataset/Copy_Titanic-Dataset.csv")
survival_by_sex_embarked = df.groupby(['Embarked', 'Sex'])['Survived'].mean().unstack()
print(survival_by_sex_embarked)
```

```python
print(df.columns)
```

```python
ports = survival_by_sex_embarked.index
x = range(len(ports))
width = 0.35
plt.bar([p - width/2 for p in x], survival_by_sex_embarked['female'], width=width, color='pink', label='Female')
plt.bar([p + width/2 for p in x], survival_by_sex_embarked['male'], width=width, color='lightblue', label='Male')
plt.xticks(x, ports)
plt.xlabel("Port of Embarkation")
plt.ylabel("Mean Survival Rate")
plt.title("Mean Survival Rate by Gender and Embarkation Port")
plt.ylim(0, 1)
plt.legend()
for i in x:
    plt.text(i - width/2, survival_by_sex_embarked['female'][i] + 0.02, f"{survival_by_sex_embarked['female'][i]:.2f}", ha='center', fontweight='bold')
    plt.text(i + width/2, survival_by_sex_embarked['male'][i] + 0.02, f"{survival_by_sex_embarked['male'][i]:.2f}", ha='center', fontweight='bold')
plt.show()
```
