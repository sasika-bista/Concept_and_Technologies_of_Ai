```python
time_data = [
(3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
(4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
(5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
(3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
(4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]
```

Task 1

```python
# 1) creating empty list
Low =[]
Moderate=[]
High=[]
```

```python
#2) iterate over time data
for data in time_data:
    for hour in data:
        if hour < 3:
            Low.append(hour)
        elif 3 <= hour <= 5:
            Moderate.append(hour)
        else:
            High.append(hour)
```

```python
#3) print the list
print("list for study time less tahn 3:",Low)
print("list for study time in between 3 and 5:",Moderate)
print("list for study time more tahn 5:",High)
```

```python
#1)no of days for low study time
print("the number of days for low study time is :",len(Low))
#2)no of days for moderate study time
print("the number of days for moderate study time is :",len(Moderate))
#3)no of days for high study time
print("the number of days foe high study time is :",len(High))
```

```python
#1)
study_list=[]
entertainment_list=[]
sleep_list=[]
```

```python
#2)
for study_hours, entertainment_hours, sleep_hours in time_data:
    study_list.append(study_hours)
    entertainment_list.append(entertainment_hours)
    sleep_list.append(sleep_hours)
```

```python
#3)
def getAvg(hours):
  sum =0
  count =0
  for hour in hours:
    sum += hour
    count += 1
  avg = (sum/count)
  return avg
print("total study hours is:",getAvg(study_list))
print("total entertainment hours is:",getAvg(entertainment_list))
print("total sleep hours is:",getAvg(sleep_list))
```

```python
import matplotlib.pyplot as plt
plt.scatter(study_list, sleep_list, color='blue')
plt.xlabel("Study Hours")
plt.ylabel("Sleep Hours")
plt.title("Scatter Plot: Study Hours vs Sleep Hours")
plt.show()
```

```python
del min
def min_coins(coins, amount):
    dp = [float("inf")] * (amount + 1)
    dp[0] = 0
    for coin in coins:
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)
    return dp[amount] if dp[amount] != float("inf") else -1

coins = [1, 2, 5]
amount = 11
print(min_coins(coins, amount))
```

```python
def longest_common_subsequence(s1, s2):
    m = len(s1)
    n = len(s2)
    dp = [[0] * (n + 1) for _ in range(m + 1)]
    for i in range(1, m + 1):
        for j in range(1, n + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])
    return dp[m][n]
s1 = "abcde"
s2 = "ace"
print("Length of LCS:", longest_common_subsequence(s1, s2))
```
