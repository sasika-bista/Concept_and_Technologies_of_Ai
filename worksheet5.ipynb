```python
def test_train_split(x,y,test_size=0.3,randomSeed =42):
  np.random.seed(randomSeed)
  indicies = np.arange(x.shape[0])
  np.random.shuffle(indicies)

  test_split_size = int(len(x) * test_size)

  test_indicies = indicies[:test_split_size]
  train_indicies = indicies[test_split_size:]

  x_test, x_train = x[test_indicies],x[train_indicies]
  y_test, y_train = y[test_indicies],y[train_indicies]

  return x_test,x_train,y_train,y_test
```

```python
def cost_function(x,y,w):
    x_array = np.array(x,dtype=float)
    y_array =np.array( y, dtype=float).reshape(-1,1)
    w = np.array(w).reshape(-1,1)
    Y_pred = np.dot(x_array,w)
    print(Y_pred.shape)
    error = Y_pred - y_array
    cost = (1 / x_array.shape[1]) * np.sum(error ** 2)
    return cost
```

```python
def gradient_descent(x,y,w,alpha=0.01,iteration=10):
  x_array = np.array(x,dtype=float)
  y_array =np.array( y, dtype=float).reshape(-1,1)
  w = np.array(w).reshape(-1,1)
  # print(w)
  cost_history = [0]*iteration
  m=len(y)
  w_update =w.copy()
  for iteration in range(iteration):
    # print("inloop",w.shape)
    y_pred = np.dot(x_array,w)
    loss = y_pred - y_array
    # print("loss",loss.shape)
    # print("X",x_array.shape)
    dw = (1/m)*sum(x_array.T @ loss)
    w_update = w_update - alpha * dw
    cost = cost_function(x,y,w_update)
    cost_history[iteration] = cost
    print("cost ",cost_history[iteration])

  return w_update, cost_history


```

```python
np.random.seed(0) # For reproducibility
X = np.random.rand(100, 3) # 100 samples, 3 features
Y = np.random.rand(100)
W = np.random.rand(3) # Initial guess for parameters
# print("in ",w)
# Set hyperparameters
alpha = 0.01
iterations = 10
# Test the gradient_descent function
final_params, cost_history = gradient_descent(X, Y, W, alpha, iterations)
# Print the final parameters and cost history
print("Final Parameters:", final_params)
print("Cost History:", cost_history)
```
